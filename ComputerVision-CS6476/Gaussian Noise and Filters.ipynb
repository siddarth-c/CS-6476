# Gaussian Noise and Filters 

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
img=cv2.imread('001.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

### Create noise to be added with the image

```python
width = img_gray.shape[0]
guas_filter = np.random.randn(width,width) * 25
plt.imshow(guas_filter,cmap='gray')
plt.show()
```

```python
comb = img_gray + guas_filter
plt.imshow(comb,cmap='gray')
plt.show()
```

### Keeping sigma constant (=10), vary the filter size to see effect.

```python
figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')
for k in range(5):
    i = pow(3,k)
    blur = cv2.GaussianBlur(comb,(i,i),10)
    plt.subplot(1,5,k+1);plt.title(i);plt.imshow(blur)
```

More filter size = More area covered to calculate average = More Blurry

### Keeping filter size constant ( =(101,101) ), vary sigma to see effect

```python
figure(figsize=(20, 20))
for k in range(5):
    i = pow(3,k)
    blur = cv2.GaussianBlur(comb,(101,101),i)
    plt.subplot(1,5,k+1);plt.title(i);plt.imshow(blur)
```

More sigma = Less sharp the bell curve = Giving the neighbour pixels more weights = More Blurry
