# Edge Detections

```python
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
img = cv.imread('SANDYandBALAJEE.jpg',cv.IMREAD_GRAYSCALE)
img_blur = cv.GaussianBlur(img,(15,15),5)

figure(figsize=(10, 10))
plt.imshow(img_blur,cmap = 'gray')
plt.show()
```

```python
img_blur.shape
```

## Method 1 - Shift image and subtract

### Y Gradient 

```python
img1 = img_blur.copy()
img1[0:img_blur.shape[0]-1] = img_blur[1:img_blur.shape[0]]

figure(figsize=(10, 10))
plt.imshow(img1-img_blur,cmap = 'gray')
plt.show()
```

### X Gradient

```python
img2 = img_blur.copy()
img2[:,0:img_blur.shape[1]-1] = img_blur[:,1:img_blur.shape[1]]

figure(figsize=(10, 10))
plt.imshow(img2-img_blur,cmap = 'gray')
plt.show()
```

## Method 2 - Using Filters and correlation

#### cv.Canny(img,low_thresh,upper_thresh)

low_thresh - Edges below this level are discarded (weak edges) <br>
upper_thresh - Edges above this level are selected (strong edges)<br>

Edges between these 2 thresholds are checked whether they are connected with strong edges. If connected then selected else discarded. <br>
https://www.youtube.com/watch?v=sRFM5IEqR2w - Computerphile on Canny filters

```python
img3 = cv.Canny(img,90,180)

figure(figsize=(10, 10))
plt.imshow(img3,cmap = 'gray')
plt.show()
```
